```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('bmh')
```

```python
x = np.array([10, 20, 30, 40, 50, 60])
y = x**2 + 20*np.random.normal(size=6)
```

```python
plt.plot(x, y, '-o')
plt.xlabel('Time')
plt.ylabel('Money')
```

```python
plt.savefig('example_figure.pdf', pad_inches=0, bbox_inches='tight')
```
